```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
```

```python
url = pd.read_excel('/content/Real estate valuation data set.xlsx')
df = url
df
```

```python
data = df[ df['X3 distance to the nearest MRT station'] <= 500 ]
data = data[ data['X4 number of convenience stores'] <= 7 ]
print(len(data))
print(data.size)
age = data['X2 house age']
price = data['Y house price of unit area']
x = age.values
y = price.values.ravel()
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)
a = model.coef_[0]
b = model.intercept_
Ynew = a*x + b
plt.scatter(age, price)
plt.plot(x, Ynew)
```
