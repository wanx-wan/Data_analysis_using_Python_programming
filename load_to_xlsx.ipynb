```python
import os
import pandas as pd
folder_path = "/Users/wanc/Desktop/CloudComp/PM2.5"

result_df = pd.DataFrame()

for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_excel(file_path, engine='openpyxl')
        if "Date" in df.columns and "54T" in df.columns:
            df["Date"] = pd.to_datetime(df["Date"]).dt.date
            df = df.dropna()
            df = df[["Date", "54T"]]
            result_df = pd.concat([result_df, df], ignore_index=True)
result_df.to_excel("/Users/wanc/Desktop/CloudComp/data_pm_54T.xlsx", index=False, engine='openpyxl')
```

```python
import os
import pandas as pd
from dateutil.parser import parse

def parse_thai_date(date_str):
    thai_months = {
        'มกราคม': 'January',
        'กุมภาพันธ์': 'February',
        'มีนาคม': 'March',
        'เมษายน': 'April',
        'พฤษภาคม': 'May',
        'มิถุนายน': 'June',
        'กรกฎาคม': 'July',
        'สิงหาคม': 'August',
        'กันยายน': 'September',
        'ตุลาคม': 'October',
        'พฤศจิกายน': 'November',
        'ธันวาคม': 'December'
    }
    
    parts = date_str.split()
    day = int(parts[0])
    month = thai_months[parts[1]]
    year = int(parts[2]) - 543
    
    return parse(f"{year}-{month}-{day}", dayfirst=True)

file_path = "/Users/wanc/Desktop/CloudComp/Sirindhorn_Dam.xlsx"
df = pd.read_excel(file_path)
df = df.iloc[2:-2]
df.rename(columns={"Unnamed: 1": "Date", "Unnamed: 8": "Quantity"}, inplace=True)
df['Date'] = df['Date'].apply(lambda x: parse_thai_date(x) if isinstance(x, str) else None)
df.dropna(subset=['Date'], inplace=True)
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
columns_to_drop = ['ตารางสรุปสภาพน้ำใน เขื่อนสิรินธร ภาคตะวันออกเฉียงเหนือ ', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 
                   'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11']
df.drop(columns=columns_to_drop, inplace=True)
df = df.iloc[3:]
df.to_excel("/Users/wanc/Desktop/CloudComp/Sirindhorn_Data.xlsx", index=False, engine='openpyxl')
file_path = "/Users/wanc/Desktop/CloudComp/Sirindhorn_Data.xlsx"
df = pd.read_excel(file_path)
print(df.columns)
df
```
